```python
import matplotlib.pyplot as plt
import numpy as np

class Lotka_Volterra():
    
    def __init__(self,a,b,d,g,max_time=10,dt=0.001,x=20,y=1):
        
        self.max_time = max_time
        self.dt = dt
        
        x_graph = []
        y_graph = []
    
        i = 0
        steps = max_time/dt
    
        while i < steps:
    
            x_change = a*x-b*x*y   #set of differential equations
            y_change = d*x*y-g*y
    
            x = x + x_change*dt
            y = y + y_change*dt
    
            x_graph.append(x)
            y_graph.append(y)
        
            i += 1
            
        self.x_graph = x_graph
        self.y_graph = y_graph
    
    def Zeit(self):
    
        t = np.arange(0,self.max_time,self.dt)
    
        plt.plot(t,self.x_graph, label = "prey")
        plt.plot(t,self.y_graph, label = "predator") 
        plt.legend()
        plt.show()
    
    def Gegeneinander(self):
        
        plt.plot(self.x_graph,self.y_graph, color = "green")
        plt.show()
        
```

a --> Vermehrungsfaktor Beute 
b --> Wahrscheinlichkeit erfolgreicher Riss
d --> Wahrscheinlichkeit erfolgreiche Ernährung
g --> Sterbefaktor Jäger

```python
def Lotka_Volterra_func(a,b,d,g,max_time=10,dt=0.001,x=20,y=1):
    
        x_graph = []
        y_graph = []
    
        i = 0
        steps = max_time/dt
    
        while i < steps:
    
            x_change = a*x-b*x*y - 11*x
            y_change = d*x*y-g*y - 0.0001*y
    
            x = x + x_change*dt 
            y = y + y_change*dt 
    
            x_graph.append(x)
            y_graph.append(y)
        
            i += 1
        
        t = np.arange(0,max_time,dt)
    
        plt.plot(t,x_graph, label="Beute")
        plt.plot(t,y_graph, label="Jäger")
        plt.legend()
        plt.show()
```
